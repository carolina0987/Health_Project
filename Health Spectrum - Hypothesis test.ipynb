```python
import pandas as pd
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
project = pd.read_csv("Project6.csv")
project
```

```python
project = project.drop("Unnamed: 0", axis=1)
project
```

```python
mean = project["Life expectancy (years)"].mean()
mean
```

## Hypothesis 1: life expectancy & Universal Health Coverage

```python
# H0: sample of life expectancies for cases where Universal Health Coverage is "Yes".
# H1: this sample represents life expectancies for cases where Universal Health Coverage is "No".

sample_1 = project[project["Universal Health Coverage (UHC)"] == "Yes"]["Life expectancy (years)"]
sample_2 = project[project["Universal Health Coverage (UHC)"] == "No"]["Life expectancy (years)"]


t_statistic, p_value = st.ttest_ind(sample_1, sample_2, equal_var = False, alternative = "greater")

alpha = 0.02
if p_value < alpha:
    print("Reject the null hypothesis. There is significant evidence that Universal Health Coverage is associated with a higher life expectancy.")
else:
    print("Fail to reject the null hypothesis. There is no significant evidence of a difference in life expectancy based on Universal Health Coverage status.")
    
print(t_statistic)
print(p_value)
```

## Hypothesis 2: life expectancy & obesity

```python
# H0: sample of life expectancies for cases where obesity prevalence is higher than or equal to the mean(of obesity).
# H1: this sample represents life expectancies for cases where obesity prevalence is lower than the mean(of obesity)
obesity_mean = project["Obesity among adults (BMI >= 30) %"].mean()

sample_1 = project[project["Obesity among adults (BMI >= 30) %"] >= obesity_mean]["Life expectancy (years)"]
sample_2 = project[project["Obesity among adults (BMI >= 30) %"] < obesity_mean]["Life expectancy (years)"]

t_statistic, p_value = st.ttest_ind(sample_1, sample_2, equal_var = False, alternative = "greater")

alpha = 0.02

if p_value < alpha:
    print("Reject the null hypothesis. There is significant evidence that obesity is associated with lower life expectancy.")
else:
    print("Fail to reject the null hypothesis. There is no significant evidence of a difference in life expectancy based on obesity.")
    
print(t_statistic)
print(p_value)
```

## Hypothesis 3: life expectancy & consumption of alcohol

```python
# H0: sample of life expectancies for cases where alcohol consumption is higher than or equal to the mean (of alcohol).
# H1: this sample represents life expectancies for cases where alcohol consumption is lower than the mean.
alcohol_mean = project["Consumption in liters of pure alcohol"].mean()

sample_1 = project[project["Consumption in liters of pure alcohol"] >= alcohol_mean]["Life expectancy (years)"]
sample_2 = project[project["Consumption in liters of pure alcohol"] < alcohol_mean]["Life expectancy (years)"]

t_statistic, p_value = st.ttest_ind(sample_1, sample_2, equal_var = False, alternative = "greater")

alpha = 0.02

if p_value < alpha:
    print("Reject the null hypothesis. There is significant evidence that alcohol consumption is associated with lower life expectancy.")
else:
    print("Fail to reject the null hypothesis. There is no significant evidence of a difference in life expectancy based on alcohol consumption.")
    
print(t_statistic)
print(p_value)
```
